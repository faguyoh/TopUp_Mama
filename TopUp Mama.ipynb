```python
customers = pd.concat([kcust, ncust], axis=0)
deliveries = pd.concat([kdeliv, ndeliv], axis=0)
orders = pd.concat([korders, norders], axis=0)
```

```python
data = pd.concat([tr, orders], axis=0)
```
